```python
%matplotlib inline
import skimage.io
import skimage.morphology
import matplotlib.pyplot as plt

import os.path

import numpy as np
```

```python
annot_dir = '/home/jr0th/github/segmentation/data/BBBC022_hand_200/raw_annotations/'
label_dir = '/home/jr0th/github/segmentation/experiments/BBBC022_hand_200/GT_labels/'
images_dir = '/home/jr0th/github/segmentation/data/BBBC022_hand_200/test/x_big/'
```

# GENERATE FILE LIST

```python
filelist = []
for root, dirs, files in os.walk(images_dir):
    filelist = filelist + files    
```

```python
for filename in filelist:
    
    annot = skimage.io.imread(annot_dir + filename)[:,:,0]
    print(np.unique(annot))
    print(annot.dtype)
    plt.imshow(annot)
    plt.show()
    
    label = skimage.morphology.label(annot)
    plt.imshow(label)
    plt.show()

    skimage.io.imsave(label_dir + filename, label)
```
