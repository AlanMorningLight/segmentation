```python
%matplotlib inline
import skimage.io
import skimage.morphology
import matplotlib.pyplot as plt

import os.path
```

```python
pred_dir = '/home/jr0th/github/segmentation/data/BBBC022_hand/all_labels/'
dir_label = '/home/jr0th/github/segmentation/experiments/BBBC022_hand/GT_labels/'
```

```python
preds = skimage.io.imread_collection(pred_dir + "*.png")

for index in range(len(preds)):

    path = preds.files[index]
    filename = os.path.basename(path)
    
    pred = preds[index]
    cell = pred == 1

    label = skimage.morphology.label(cell)
    skimage.io.imsave(dir_label + filename, label)
```
