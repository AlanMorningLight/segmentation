```python
%matplotlib inline
import matplotlib.pyplot as plt

import sys
sys.path.append('/home/jr0th/github/segmentation/code/')

import skimage.io
import skimage.morphology
import skimage.segmentation
import sklearn.metrics

import os.path
import os

import numpy as np

import time

debug = True
```

```python
data_dir = '/home/jr0th/github/segmentation/code/analysis/data/'
all_gt_files = os.listdir(data_dir + 'y_gt')
```

```python
def probmap_to_pred(probmap):
    pred = np.zeros((probmap.shape[0], probmap.shape[1]), np.uint8)
    
    # we need to boost the boundary class to make it more visible
    # this shrinks the cells a little bit but avoids undersegmentation
    boundary_boost_factor = 100
    argmax = np.argmax(probmap * [1, 1, boundary_boost_factor], 2)
    
    pred[argmax == 0] = 0 # background
    pred[argmax == 1] = 1 # cell
    pred[argmax == 2] = 255 # boundary
    
    return pred
```

```python
def pred_to_label(pred):
    
    cell = (pred == 1)    
    
    # fix cells
    cell = skimage.morphology.remove_small_holes(cell, min_size=300)
    cell = skimage.morphology.remove_small_objects(cell, min_size=300)
    
    # label cells only
    [label, num] = skimage.morphology.label(cell, return_num=True)
    
    if debug:
        print(label.dtype)
        print(np.unique(label))
        print(num)
        
        plt.figure(figsize=(15,15))
        plt.imshow(label)
        plt.show()
    
    return label
```

```python
# create segmentations
for i in range(len(all_gt_files)):
    
    # get filename
    gt_filename = all_gt_files[i]
    filename = os.path.splitext(gt_filename)[0]
    
    # load probmap and gt
    probmap = np.load(data_dir + 'y_probmap/' + filename + '.npy')
    gt = skimage.io.imread(data_dir + 'y_gt/' + gt_filename)

    # create prediction from probability map
    pred = probmap_to_pred(probmap)
    
    # create labels from predictions
    label_gt = pred_to_label(gt)
    label_pred = pred_to_label(pred)
    
    skimage.io.imsave(data_dir + 'seg_gt/' + filename + '.png', label_gt)
    skimage.io.imsave(data_dir + 'seg_pred/' + filename + '.png', label_pred)
```

```python
# calculate segmentation error
for i in range(len(all_gt_files)):
    
    # get filename
    gt_filename = all_gt_files[i]
    filename = os.path.splitext(gt_filename)[0]
    
    # load segmentations
    seg_pred = skimage.io.imread(data_dir + 'seg_pred/' + filename + '.png')
    seg_gt = skimage.io.imread(data_dir + 'seg_gt/' + filename + '.png')
    
    if debug:
        figure, axes = plt.subplots(ncols=2, nrows=1, figsize=(10, 5))
        axes[0].imshow(seg_pred)
        axes[0].set_title('prediction')
        axes[1].imshow(seg_gt)
        axes[1].set_title('ground truth')
        plt.show()
    
    nb_nuclei_gt = np.max(seg_gt)
    nb_nuclei_pred = np.max(seg_pred)

    print('nuclei in prediction: ' + str(nb_nuclei_pred))
    print('nuclei in GT: ' + str(nb_nuclei_gt))
    
    IoUs = np.full((nb_nuclei_gt, nb_nuclei_pred), -1, dtype = np.float32)

    # calculate IoU for each nucleus index_gt in GT and nucleus index_pred in prediction    
    # TODO improve runtime of this algorithm
    for index_gt in range(1,nb_nuclei_gt+1):
        nucleus_gt = seg_gt == index_gt
        number_gt = np.sum(nucleus_gt)
        for index_pred in range(1,nb_nuclei_pred+1):
            nucleus_pred = seg_pred == index_pred            
            number_pred = np.sum(nucleus_pred)
            
            same_and_1 = np.sum((nucleus_gt == nucleus_pred) * nucleus_gt)
            
            IoUs[index_gt-1,index_pred-1] = same_and_1 / (number_gt + number_pred - same_and_1)

    plt.figure(figsize=(10,10))
    plt.matshow(IoUs, cmap=plt.cm.gray)
    plt.show()
    
    np.save(data_dir + 'mat/' + filename, IoUs)
    
```

```python
# calculate splits and merges
for i in range(len(all_gt_files)):
    
    # get filename
    gt_filename = all_gt_files[i]
    filename = os.path.splitext(gt_filename)[0]
    
    mat = np.load(data_dir + 'mat/' + filename + '.npy')
    
    nb_nuc_gt = mat.shape[0]
    nb_nuc_pred = mat.shape[1]
    
    if debug:
        print('# nuclei gt', nb_nuc_gt)
        print('# nuclei pred', nb_nuc_pred)
    
    # only allow assignments if IoU is at least 0.5
    detection_map = (mat > 0.5)
    detection_map_gt = np.sum(detection_map, 1)
    detection_map_pred = np.sum(detection_map, 0)
    
    detection_rate = mat * detection_map
    
    nb_matches = np.sum(detection_map)
    
    if debug:
        print('# matches', nb_matches)
        
    nb_overdetection = nb_nuc_pred - nb_matches
    nb_underdetection = nb_nuc_gt - nb_matches
    
    if debug:
        print('# overdetections', nb_overdetection)
        print('# underdetections', nb_underdetection)
        
    mean_IoU_gt = np.mean(np.sum(detection_rate, axis = 1))
    
    if debug:
        print('# mean IoU of gt nuclei', mean_IoU_gt)
    
    plt.figure(figsize=(10,10))
    plt.matshow(detection_rate)
    plt.show()
    
    # visualize error zones
    # load segmentations
    seg_pred = skimage.io.imread(data_dir + 'seg_pred/' + filename + '.png')
    seg_gt = skimage.io.imread(data_dir + 'seg_gt/' + filename + '.png')
    
    
    error_gt = np.nonzero(detection_map_gt == 0)[0] + 1
    error_pred = np.nonzero(detection_map_pred == 0)[0] + 1
    print(error_gt)
    print(error_pred)
    error_img = np.zeros((seg_pred.shape[0], seg_pred.shape[1], 3))

    for error_nuc_gt_id in error_gt:
        error_img[seg_gt == error_nuc_gt_id, 0] = 100
    for error_nuc_pred_id in error_pred:
        error_img[seg_pred == error_nuc_pred_id, 1] = 100
        
    figure, axes = plt.subplots(ncols=3, nrows=1, figsize=(30, 10))
    axes[0].imshow(seg_pred)
    axes[0].set_title('prediction')
    axes[1].imshow(seg_gt)
    axes[1].set_title('ground truth')
    axes[2].imshow(error_img)
    axes[2].set_title('errors')
    plt.show()
    
    print(error_gt)
```
